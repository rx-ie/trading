## Setting up API Keys in Colab Secrets

To enable Serp API and Alpha Vantage functionalities, you need to obtain API keys for these services and store them securely in Colab Secrets. If you haven't already, please do the following:

1.  **Gemini API Key**: Add your Gemini API key to Colab Secrets (the "🔑" icon in the left panel) with the name `GEMINI_API_KEY`.
2.  **Serp API**: Get your API key from [serpapi.com](https://serpapi.com/) and add it to Colab Secrets with the name `SERPAPI_API_KEY`.
3.  **Alpha Vantage**: Get your API key from [alphavantage.co](https://www.alphavantage.co/free/) and add it to Colab Secrets with the name `ALPHA_VANTAGE_API_KEY`.

These keys will be loaded from your secrets for secure use.

```python
import os
from google.colab import userdata
from tradingagents.graph.trading_graph import TradingAgentsGraph
from tradingagents.default_config import DEFAULT_CONFIG

# Load API keys from Colab Secrets and set them as environment variables
os.environ["GEMINI_API_KEY"] = userdata.get('GEMINI_API_KEY')
os.environ["SERPAPI_KEY"] = userdata.get('SERPAPI_KEY')
os.environ["ALPHA_VANTAGE_KEY"] = userdata.get('ALPHA_VANTAGE_KEY')

# Start with a copy of the default configuration
config = DEFAULT_CONFIG.copy()

# Configure for Gemini LLM
config["llm_provider"] = "google"
config["deep_think_llm"] = "models/gemini-pro-latest" # Use a suitable Gemini model
config["quick_think_llm"] = "models/gemini-flash-latest" # Use a suitable Gemini model
config["google_api_key"] = os.environ["GEMINI_API_KEY"] # Explicitly set Google API key

# Enable and configure Serp API
config["enable_serpapi"] = True
config["serpapi_api_key"] = os.environ["SERPAPI_KEY"]

# Enable and configure Alpha Vantage
config["enable_alpha_vantage"] = True
config["alpha_vantage_api_key"] = os.environ["ALPHA_VANTAGE_KEY"]

# Disable other providers as per your request to only use the specified ones
config["enable_yahoo_finance"] = False
config["enable_fred"] = False
config["enable_twitter_data"] = False

# Initialize the agents with the custom configuration
ta_new_notebook = TradingAgentsGraph(debug=True, config=config)

print("TradingAgentsGraph initialized for new notebook with custom configuration.")
print("Enabled services:")
print(f"  LLM Provider: {config['llm_provider']}")
print(f"  Serp API: {config['enable_serpapi']}")
print(f"  Alpha Vantage: {config['enable_alpha_vantage']}")
```

## Run a Sample Analysis

Now, you can run an analysis using the `ta_new_notebook` instance, which is configured to use Gemini, Serp API, and Alpha Vantage. Be mindful of API call limits with free tiers for Alpha Vantage and Serp API.

```python
# Example: Run an analysis for a specific stock and date
try:
    # Using 'U96.SI' as an example ticker and a specific date
    _, decision_result = ta_new_notebook.propagate("U96.SI", "2026-07-03")
    print("\n--- New Notebook Analysis Result ---")
    print(decision_result)
except Exception as e:
    print(f"An error occurred during analysis: {e}")
    print("Please ensure all API keys are correctly set in Colab Secrets and check your free tier usage limits.")
```
